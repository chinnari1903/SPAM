.....STANDARIZATION

```python
df["stanfdarized_Avg_Price"]=(df["Avg Price"]-df["Avg Price"].mean())/df["Avg Price"].std()
```

```python
df["stanfdarized_Avg_Price"].describe()
```

```python
fig=plt.figure(figsize=(15,10))
plt.subplot(2,1,2)
plt.hist(df['Avg Price'])
plt.title('BEFORE')

plt.subplot(1,2,2)
plt.hist(df['stanfdarized_Avg_Price'])
plt.title('AFTER')
plt.show()
```
